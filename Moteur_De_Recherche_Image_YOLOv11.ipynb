```python
import os
import random
import shutil
import glob
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2  # used to read the image size for bounding-box conversion
import yaml

# 1) LOAD YOUR TRAINED MODEL
model = YOLO("/content/best.pt")

# 2) PATHS
# plastic
# new_image_path = "/content/AugTrashNet-4/test/images/plastic75_jpg.rf.fe4c3ce5343da9dd1bb62d6bb8eef0e2.jpg"

# glass 91
# new_image_path = "/content/AugTrashNet-4/test/images/glass97_jpg.rf.09ad491eccc8a4db03e834ce296be54c.jpg"

# cardboard 91
# new_image_path = "/content/AugTrashNet-4/test/images/cardboard219_jpg.rf.c955a392b9376f61ccd7c71bb2c33608.jpg"

# metal 97,95
# new_image_path = "/content/AugTrashNet-4/test/images/metal387_jpg.rf.6e1cd606fd00b5756504c66e82d48187.jpg"
new_image_path = "/content/AugTrashNet-4/test/images/metal52_jpg.rf.9d49d3dea2b6564504e0a8f9d3a22580.jpg"

# paper 96
# new_image_path = "/content/AugTrashNet-4/test/images/paper134_jpg.rf.7a312584397952d30962a507a575981e.jpg"
# new_image_path = "/content/paper-test.jpg"

confidenceScore = 0.85

folder_newImage = "/content/newImage"
folder_newImage_images = os.path.join(folder_newImage, "images")
folder_newImage_labels = os.path.join(folder_newImage, "labels")

test_images_dir  = "/content/AugTrashNet-4/test/images"
test_labels_dir  = "/content/AugTrashNet-4/test/labels"

# Ensure output folders exist
os.makedirs(folder_newImage_images, exist_ok=True)
os.makedirs(folder_newImage_labels, exist_ok=True)

# 3) PREDICT ON THE NEW IMAGE
results = model.predict(
    source=new_image_path,
    show=False,
    save=False,
    line_width=2,
    show_labels=True,
    show_conf=True
)

# 4) PARSE THE RESULT & CHECK CONFIDENCE
if len(results) == 0:
    print("No results from the model.")
    exit()

pred_boxes = results[0].boxes
if pred_boxes is None or len(pred_boxes) == 0:
    print("No objects detected in the new image.")
    exit()

# Convert BGR -> RGB for correct color in Matplotlib
annotated_img = results[0].plot()
annotated_img = annotated_img[..., ::-1]

plt.figure(figsize=(8,6))
plt.imshow(annotated_img)
plt.title("Predictions on the New Image")
plt.axis("off")
plt.show()

# Find the detection with highest confidence
max_conf   = 0.0
best_class = None

for box in pred_boxes:
    # Use box.conf.item() and box.cls.item() to avoid the deprecated "ndim > 0" warning
    conf = float(box.conf.item())
    cls  = int(box.cls.item())

    if conf > max_conf:
        max_conf   = conf
        best_class = cls

print(f"Highest confidence detection: class={best_class}, conf={max_conf:.2f}")

# 5) IF >= confidenceScore, copy new image + create a label file, else stop
if max_conf >= confidenceScore:
    base_name = os.path.basename(new_image_path)
    base_txt  = os.path.splitext(base_name)[0]  # e.g. "paper-test"

    # Copy the image into newImage/images
    dst_img_path = os.path.join(folder_newImage_images, base_name)
    shutil.copy(new_image_path, dst_img_path)
    print(f"Copied {new_image_path} to {dst_img_path} because conf >= {confidenceScore}.")

    # 5A) Create a YOLO-format label for this new image
    label_txt_path = os.path.join(folder_newImage_labels, base_txt + ".txt")

    # read original image size to convert bounding boxes
    img_cv = cv2.imread(new_image_path)
    if img_cv is None:
        print("Warning: could not read image for bounding-box conversion.")
    else:
        h, w, _ = img_cv.shape
        with open(label_txt_path, "w") as f_label:
            for box in pred_boxes:
                cls_idx = int(box.cls.item())
                conf    = float(box.conf.item())
                # box.xyxy[0] is still fine, as that array shape is (4,)
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()

                # Convert to YOLO format
                bbox_width   = x2 - x1
                bbox_height  = y2 - y1
                x_center     = x1 + bbox_width / 2
                y_center     = y1 + bbox_height / 2

                # Normalize [0..1]
                x_center_n = x_center / w
                y_center_n = y_center / h
                width_n    = bbox_width / w
                height_n   = bbox_height / h

                # Write line: "<class> <x_center> <y_center> <width> <height>"
                f_label.write(f"{cls_idx} {x_center_n:.6f} {y_center_n:.6f} {width_n:.6f} {height_n:.6f}\n")

        print(f"Created YOLO label file: {label_txt_path}")

    # 5B) Retrieve 5 random images from the test set with the same best_class
    matching_images = []
    label_files = sorted(glob.glob(os.path.join(test_labels_dir, "*.txt")))

    for label_file in label_files:
        with open(label_file, "r") as lf:
            lines = lf.readlines()
            for line in lines:
                line_split = line.strip().split()
                if len(line_split) > 0:
                    class_idx = int(line_split[0])
                    if class_idx == best_class:
                        base_txt_candidate = os.path.splitext(os.path.basename(label_file))[0]
                        img_name_jpg = base_txt_candidate + ".jpg"
                        img_name_png = base_txt_candidate + ".png"
                        cand_jpg = os.path.join(test_images_dir, img_name_jpg)
                        cand_png = os.path.join(test_images_dir, img_name_png)
                        if os.path.exists(cand_jpg):
                            matching_images.append(cand_jpg)
                        elif os.path.exists(cand_png):
                            matching_images.append(cand_png)
                        break

    print(f"Found {len(matching_images)} images in test set with class {best_class}.")

    if len(matching_images) == 0:
        print("No test images found with that class. Stopping.")
    else:
        num_to_show = min(5, len(matching_images))
        sample_imgs = random.sample(matching_images, num_to_show)

        print("Displaying these sample images from the same class:")
        for s in sample_imgs:
            print("  ", s)

        # Load class names from your YAML
        with open('/content/AugTrashNet-4/data.yaml', 'r') as file:
            data = yaml.safe_load(file)
        classes = data['names']

        fig, axes = plt.subplots(1, num_to_show, figsize=(5*num_to_show, 4))

        for i, img_path in enumerate(sample_imgs):
            img_arr = plt.imread(img_path)
            axes[i].imshow(img_arr)
            class_name = classes[best_class]
            axes[i].text(
                0.5, -0.1,
                f"{class_name} (class {best_class})",
                transform=axes[i].transAxes,
                ha='center',
                va='top',
                fontsize=12
            )
            axes[i].axis("off")

        plt.subplots_adjust(bottom=0.2)
        plt.tight_layout()
        plt.show()

else:
    print(f"Confidence < {confidenceScore*100:.0f}% ({max_conf:.2f}). Stopping the process.")
```
